```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
file_path = 'salesdaily.csv'
df = pd.read_csv(file_path)

# Set the style for visualizations
sns.set(style="whitegrid")

# Convert the 'datum' column to datetime for easier manipulation
df['datum'] = pd.to_datetime(df['datum'])

# Set the 'datum' column as the index for easier time series plotting
df.set_index('datum', inplace=True)

# Time Series Plot: Sales Over Time for Selected Product Categories
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['M01AB'], label='M01AB', color='blue')
plt.plot(df.index, df['M01AE'], label='M01AE', color='green')
plt.plot(df.index, df['N02BA'], label='N02BA', color='red')
plt.plot(df.index, df['N02BE'], label='N02BE', color='purple')

plt.title('Sales Over Time for Selected Product Categories')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Bar Plot: Total Sales for Each Product Category
plt.figure(figsize=(10, 6))
sales_columns = ['M01AB', 'M01AE', 'N02BA', 'N02BE', 'N05B', 'N05C', 'R03', 'R06']
total_sales = df[sales_columns].sum()

sns.barplot(x=total_sales.index, y=total_sales.values, palette='Set2')
plt.title('Total Sales for Each Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
